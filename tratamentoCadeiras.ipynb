```python
import pandas as pd

df = pd.read_excel('cadeiras.xlsx')

display(df)
```

```python
# Retirar o 'R$' do preco

df['preco'] = df['preco'].str.replace("R$ ", "", regex=False)

display(df)
```

```python
# transformar o preco em float

df['preco'] = df['preco'].str.replace(".", "").str.replace(",", ".").astype(float)

display(df.info())
```

```python
# padronizando o nome

df['nome'] = df['nome'].str.capitalize().str.strip()

display(df)

```

```python
# Identificando as cores

cores = [
    'amarelo', 'amarela', 'yellow',
    'azul', 'blue',
    'branco', 'branca', 'white',
    'cinza', 'gray', 'grey',
    'laranja', 'orange',
    'marrom', 'brown',
    'preto', 'preta', 'black',
    'rosa', 'pink',
    'roxo', 'roxa', 'purple',
    'verde', 'green',
    'vermelho', 'vermelha', 'red', 'prata', 'mercury',
    'ciano', 'cyan', 'uva', 'geada', 'grafite', 'silver', 'vinho', 'bege', 'alumínio', 'ouro', 'grape'
]


def achar_cor(texto):
    for cor in cores:
        if cor in texto:
            return cor
    return 'Não identificada'

df['cor'] = df['nome'].apply(achar_cor)

print(len(df[df['cor'] == 'Não identificada']))
```

```python
# Padronizando a cor 
df['cor'] = df['cor'].str.lower().str.strip()

df['cor'] = df['cor'].str.replace("preta", "preto").str.replace("branca", "branco").str.replace("vermelha", "vermelho")
df['cor'] = df['cor'].str.replace("roxa", "roxo") .str.replace("amarela", "amarelo").str.replace("blue", "azul")
df['cor'] = df['cor'].str.replace("yellow", "amarelo") .str.replace("white", "branco").str.replace("gray", "cinza")
df['cor'] = df['cor'].str.replace("orange", "laranja") .str.replace("brown", "marrom").str.replace("black", "preto")
df['cor'] = df['cor'].str.replace("pink", "rosa") .str.replace("purple", "roxo").str.replace("green", "verde")
df['cor'] = df['cor'].str.replace("red", "vermelho").str.replace("uva", "roxo").str.replace("geada", "branco")  
df['cor'] = df['cor'].str.replace("grafite", "preto").str.replace("silver", "cinza").str.replace("prata", "cinza")  
df['cor'] = df['cor'].str.replace("vinho", "vermelho").str.replace("alumínio", "cinza").str.replace("ouro", "amarelo")
df['cor'] = df['cor'].str.replace("grape", "roxo").str.replace("mercury", "cinza")  

```

```python
# identificando os 'Não identificada'
pd.set_option('display.max_colwidth', None)
display(df['nome'][df['cor'] == 'Não identificada'])
pd.reset_option('display.max_colwidth')
```
